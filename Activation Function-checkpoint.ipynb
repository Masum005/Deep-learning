```python
import numpy as np #number analysis # matrix calculator
```

```python
import matplotlib.pyplot as plt # visualisation
```

```python
%matplotlib inline 
```

```python
x = np.arange( -5, 5, 0.01)
```

```python
def plot (func, yaxis = (-1.4, 1.4)):
    plt.ylim(yaxis) #limitation (-1.4,1.4)
    plt.locator_params(nbins = 6)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.axhline (lw = 2, c = 'black')
    plt.axvline (lw = 1, c =  'green')
    plt.grid(alpha = 0.7, ls ='-.')
    plt.box (on = None)
    plt.plot (x , func(x), c = 'blue', lw = 4) #recursion = repeted
```

```python
binary_step = np.vectorize(lambda x : 1 if x > 0 else 0 , otypes = [ np.float] )
```

```python
plot(binary_step, yaxis = (-0.4, 1.4)) # constractor
```

```python
plot(binary_step)
```

```python
piecewise_linear   = np. vectorize (lambda x: 1 if x>3 else 0 if x<3 else 2)
```

```python
plot (piecewise_linear )
```

```python
def sigmoid(x):
    
    return 1/(1 + np.exp(-x))
```

```python
plot(sigmoid) # sigmoid value always (0 to 1 )
```

```python
relu =  np.vectorize ( lambda x: x if x>0 else 0, otypes =[ np.float])
```

```python
plot(relu)
```

```python
leaky_relu =  np.vectorize ( lambda x: x if max(0.1 *x, x) else 0, otypes =[ np.float])   
```

```python
plot(leaky_relu) # if there is a problem leaky relu value will help you
```

```python
elu =  np.vectorize ( lambda x: x if x>0 else 0.5* (np. exp(x) - 1), otypes =[ np.float])
```

```python
def softplus ( x):
    return np.log(1 + np.exp(x))
```

```python
plot(softplus) #always bigger than 1 
```
